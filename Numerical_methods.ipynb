```python
import random
import matplotlib.pyplot as plt
import numpy as np

def estimate_pi(num_points):
    inside_circle = 0

    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1

    pi_estimate = (inside_circle / num_points) * 4
    return pi_estimate

def estimate_variance(num_points, num_trials):
    estimates = [estimate_pi(num_points) for _ in range(num_trials)]
    variance = np.var(estimates)
    return variance

# Number of points to test
points = [100, 1000, 10000, 100000, 1000000]
pi_estimates = []
variances = []

for num_points in points:
    pi_estimate = estimate_pi(num_points)
    pi_estimates.append(pi_estimate)
    variance = estimate_variance(num_points, 100)  # Run 100 trials for each number of points
    variances.append(variance)
    print(f"Estimated π with {num_points} points: {pi_estimate}, Variance: {variance}")

# Plotting the results
plt.figure(figsize=(12, 6))

# Plot for estimated π values
plt.subplot(1, 2, 1)
plt.errorbar(points, pi_estimates, yerr=variances, fmt='-o', capsize=5, label='Estimated π')
plt.axhline(y=3.14159, color='r', linestyle='--', label='True π')
plt.xscale('log')
plt.xlabel('Number of Points')
plt.ylabel('Estimated π')
plt.title('Convergence of π Estimation using Monte Carlo Method')
plt.legend()
plt.grid(True)

# Plot for sample variance
plt.subplot(1, 2, 2)
plt.plot(points, variances, '-o', label='Sample Variance')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Points')
plt.ylabel('Sample Variance')
plt.title('Decrease of Sample Variance with Number of Points')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import math

def newton_raphson(f, df, x0, tolerance=1e-10, max_iterations=1000):
    x = x0
    for _ in range(max_iterations):
        fx = f(x)
        dfx = df(x)
        if abs(fx) < tolerance:
            return x
        x = x - fx / dfx
    raise ValueError("Newton-Raphson method did not converge")

# Function f(x) = sin(x)
def f(x):
    return math.sin(x)

# Derivative f'(x) = cos(x)
def df(x):
    return math.cos(x)

# Initial guess
x0 = 3

# Estimate π using Newton-Raphson method
pi_estimate = newton_raphson(f, df, x0)
print(f"Estimated π using Newton-Raphson method: {pi_estimate}")
```
